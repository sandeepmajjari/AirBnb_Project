### Task 1 : Importing All Dependencies

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Task 2 : Loading Datasets

```python
df = pd.read_csv("New_York_listings_2024.csv", encoding_errors='ignore')
```

### Task 4 : Data Cleaning

```python
df.isnull().sum()
```

```python
# dropping all the missing values rows
df.dropna(inplace = True)
```

```python
# Now, there is no missing values
df.isnull().sum()
```

```python
df.shape
```

```python
# dealing with duplicates rows

df.duplicated().sum()
```

```python
# Displaying the duplicate rows

df[df.duplicated()]
```

```python
# deleting all the duplicate rows

df.drop_duplicates(inplace = True)

# checking the number of duplicate rows again

df.duplicated().sum()
```

```python
# checking all the data types were correct or not

df.dtypes
```

```python
# Type casting
# changing the data types

df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

# EDA

## Task 5 : Data Analysis

#### Univariate Analysis 

```python
# identifying outliers in the price column

df1 = df[df['price'] <= 1500]
sns.boxplot(data = df1, x = 'price')
plt.title('Boxplot of Price')
plt.show()
```

```python
# price distribution

plt.figure(figsize = (6,5))
sns.histplot(data = df1, x = 'price', bins = 100)
plt.title('hist. Distribution of Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# df.columns
df.dtypes
```

```python
plt.figure(figsize = (6,3))
sns.histplot(data = df1, x = 'availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby('neighbourhood_group')['price'].mean()
```

```python
df.groupby('neighbourhood_group')['price_per_bed'].mean()
```

***Feature Engineering***

```python
# creating a new column "price_per_bed"
df['price_per_bed'] = df['price'] / df['beds']
df.head()
```

**Bi Variable Analysis :**
one variable dependency in another variable

```python
df.columns
```

```python
sns.barplot(data = df, x = 'neighbourhood_group', y = 'price', hue = 'room_type')
plt.title('Price in each Neighbourhood Group')
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(data = df1, x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
plt.title('Price vs Number of Reviews')
plt.show()
```

```python
sns.pairplot(data = df1, vars = ['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue = 'room_type')
plt.show()
```

```python
# Geographical Distribution of Airbnbs in New York City

plt.figure(figsize = (10,7))
sns.scatterplot(data = df1, x = 'longitude', y = 'latitude', hue = 'room_type')
plt.title('Geographical Distribution of AirBnb in NYC')
plt.show()
```

```python
df.dtypes
```

```python
# heatmap - correlation of one variable with others for numerical columns

corr = df1[['latitude', 'longitude', 'price', 'minimum_nights', 'beds', 'number_of_reviews', 'reviews_per_month', 'availability_365']].corr()
corr
plt.figure(figsize = (8,6))
sns.heatmap(data = corr, annot = True)
plt.title('Correlation Heatmap')
plt.show()
```
